```python
# 1506 - 1512 referencing past homework
# Question 3

import sys

def maxOccurrences(inp):
    lst = inp.split(" ")
    lst = [int(i) for i in lst]
    count_dict = {}
    for i in lst:
        if count_dict.get(i) == None:
            count_dict.update({i : lst.count(i)})
    return sorted([x for x in count_dict.keys() if count_dict[x] == max(count_dict.values())]),max(count_dict.values())
```

```python
# 1518 - # 1527
class Avatar:
    def __init__(self, name, hp = 100, position = (1,1)):
        self.name = name
        self.hp = hp
        self.position = position
        
    def getName(self):
        return self.name
    
    def setName(self, name):
        self.name = name
    
    def getHP(self):
        return self.hp
    
    def setHP(self, hp):
        self.hp = hp
    
    def getPosition(self):
        return self.position
    
    def setPosition(self, position):
        self.position = position
        
    def heal(self, healing=1):
        if healing < 0:
            healing = 0
        self.hp += healing
    
    def attacked(self, damage = -1):
        if damage > 0:
            damage = 0
        self.hp += damage
```

```python
# 1527 - 1554
import copy

class Map():
    def __init__(self, world):
        self.world = copy.deepcopy(world)
    
    def whatIsAt(self, position):
        try:
            item = self.world[position]
        except:
#           print("ex getEnemyAttack")
            return 'Empty'
        if item == 'x':
            return 'Exit'
        elif item == 0:
            return 'Wall'
        elif item > 0:
            return 'Food'
        elif item < 0:
            return 'Enemy'
        else:
            return 'Empty'
        
    def getEnemyAttack(self, position):
        try:
            item = self.world[position]
        except:
#           print("ex getEnemyAttack")
            return False
        item = self.world[position]
        if item < 0:
            return item
        else:
            return False
        
    def getFoodEnergy(self, position):
        try:
            item = self.world[position]
        except:
#             print("ex getFoodEnergy")
            return False
        if item > 0:
            return item
        else:
            return False
        
    def removeEnemy(self, position):
        try:
            item = self.world[position]
        except:
#             print("ex removeEnemy")
            return False
        if item < 0:
            del self.world[position]
            return True
        else:
            return False
        
    def eatFood(self, position):
        try:
            item = self.world[position]
        except:
#             print("ex eatFood")
            return False
        if item > 0:
            del self.world[position]
            return True
        else:
            return False
        
        
    def getExitPosition(self):
        # https://stackoverflow.com/questions/8023306
        # returns dict only with keys with values that are 'x'
        exit_dict = {k:v for k, v in self.world.items() if v == 'x'}
        for key in exit_dict:
            return key
```

```python
# waste of time should give up on this question
# this is the result from 1527 - 1639
# this is from Edmund

class DW2Game(sm.SM):
    
    def __init__(self, A, M):
        self.state = copy.deepcopy((A,M))
        self.start_state = copy.copy(self.state)
        
    def done(self, state):
        a, m = state
        return a.getPosition() == m.getExitPosition()
    
    def get_next_values(self, state_, inp):
        cmd, facing = inp
        state = copy.deepcopy(state_)
        a, m = state
        
        ## We create a dictionary of possible moves ##
        moveto = {'up': (0, -1), 'down': (0, 1), 'left': (-1, 0), 'right': (1, 0)}[facing]
        ## Add them to position to simplify the move condition ##
        p = a.getPosition()
        p_topos = tuple([c1 + c2 for c1, c2 in zip(p, moveto)])
        
        whatIs = m.whatIsAt(p_topos)
        ## Instead of coding for direction, we code for action ##
        if cmd == 'move':
            
            if whatIs == 'Wall':
                pass
            
            elif whatIs == 'Enemy':
                a.attacked(m.getEnemyAttack(p_topos))
                
            elif whatIs == 'Empty':
                a.setPosition(p_topos)
            
            elif whatIs == 'Food':
                a.setPosition(p_topos)
                a.heal(m.getFoodEnergy(p_topos))
                m.eatFood(p_topos)
                
            elif whatIs == 'Exit':
                a.setPosition(p_topos)
                
        elif cmd == 'attack':
            
            if whatIs == 'Enemy':
                m.removeEnemy(p_topos)
                
        output = (a.name, a.position, a.hp)
        
        return state, output
```
